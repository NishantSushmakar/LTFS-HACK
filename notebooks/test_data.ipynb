```python
import pandas as pd
df=pd.read_excel('test_Data.xlsx')
df
```

```python
df.isnull().sum()
```

```python
df['SEX'].value_counts()
df['SEX'].fillna(df['SEX'].mode()[0], inplace=True)
```

```python
df['AGE'].fillna(df['AGE'].median(), inplace=True)
```

```python
df['MonthlyIncome'].fillna(df['MonthlyIncome'].median(), inplace=True)
```

```python
df['ZiPCODE'].fillna(df['ZiPCODE'].median(), inplace=True)
```

```python
df['ManufacturerID'].value_counts()
df['ManufacturerID'].fillna(df['ManufacturerID'].median(), inplace=True)
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['ZiPCODE']= le.fit_transform(df['ZiPCODE'])
df
```

```python
columns = ['Frequency','InstlmentMode','LoanStatus','PaymentMode','SEX']
for col in columns: 
    dummy=pd.get_dummies(df[col])
    df=pd.concat([df,dummy],axis=1)
    df.drop(col, inplace=True,axis=1)
df
```

```python
df.info()
```

```python
df=df.drop(['DisbursalDate','MaturityDAte','AuthDate'],axis=1)
```

```python
df.to_csv('test_data1.csv')
```
