```python
import pandas as pd
data_train=pd.read_excel('train_Data.xlsx')
data_test=pd.read_excel('test_Data.xlsx')
```

```python
data_train.isnull().sum()
```

```python
data_train['SEX'].fillna(data_train['SEX'].mode()[0], inplace=True)
data_train['AGE'].fillna(data_train['AGE'].median(), inplace=True)
data_train['MonthlyIncome'].fillna(data_train['MonthlyIncome'].median(), inplace=True)
data_train['ZiPCODE'].fillna(data_train['ZiPCODE'].median(), inplace=True)
```

```python
data_test.isnull().sum()
```

```python
data_test['SEX'].fillna(data_test['SEX'].mode()[0], inplace=True)
data_test['AGE'].fillna(data_test['AGE'].median(), inplace=True)
data_test['MonthlyIncome'].fillna(data_test['MonthlyIncome'].median(), inplace=True)
data_test['ZiPCODE'].fillna(data_test['ZiPCODE'].median(), inplace=True)
data_test['ManufacturerID'].fillna(data_test['ManufacturerID'].median(), inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
data_train['ZiPCODE']= le.fit_transform(data_train['ZiPCODE'])
data_test['ZiPCODE']= le.fit_transform(data_test['ZiPCODE'])
```

```python
columns = ['Frequency','InstlmentMode','LoanStatus','PaymentMode','SEX']
for col in columns: 
    dummy=pd.get_dummies(data_train[col])
    data_train=pd.concat([data_train,dummy],axis=1)
    data_train.drop(col, inplace=True,axis=1)
```

```python
columns = ['Frequency','InstlmentMode','LoanStatus','PaymentMode','SEX']
for col in columns: 
    dummy=pd.get_dummies(data_test[col])
    data_test=pd.concat([data_test,dummy],axis=1)
    data_test.drop(col, inplace=True,axis=1)
```
