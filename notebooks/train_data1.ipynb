```python
import pandas as pd
df=pd.read_excel('train_Data.xlsx')
df
```

```python
df.isnull().sum()
```

```python
df['SEX'].fillna(df['SEX'].mode()[0], inplace=True)
```

```python
df['AGE'].fillna(df['AGE'].median(), inplace=True)
df['AGE'].value_counts()
```

```python
df['MonthlyIncome'].fillna(df['MonthlyIncome'].median(), inplace=True)
df['MonthlyIncome'].value_counts()
```

```python
df.isnull().sum()
```

```python
df['ZiPCODE'].fillna(df['ZiPCODE'].median(), inplace=True)
df['ZiPCODE'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df['ZiPCODE']= le.fit_transform(df['ZiPCODE'])
df
```

```python
dummy=pd.get_dummies(df['Frequency'])
dummy.head()
```

```python
df=pd.concat([df,dummy],axis=1)
df.head()
```

```python
df.drop('Frequency', inplace=True,axis=1)
```

```python
df
```

```python
columns = ['InstlmentMode','LoanStatus','PaymentMode','SEX']
for col in columns: 
    dummy=pd.get_dummies(df[col])
    df=pd.concat([df,dummy],axis=1)
    df.drop(col, inplace=True,axis=1)
```

```python
target_dict = { 'No Top-up Service' : 0 ,
        ' > 48 Months': 1,
        '36-48 Months':2,
        '24-30 Months' : 3,
        '30-36 Months' : 4,
        '18-24 Months': 5,
        '12-18 Months': 6 
        }
target_dict
```

```python
df=df.replace({"Top-up Month" : target_dict})
df['Top-up Month']
```
